With numeric data (distribution)

```python
max = data['NUMBER OF PERSONS INJURED'].max()
min = data['NUMBER OF PERSONS INJURED'].min()
print([min, max])
```

```python
data['NUMBER OF PERSONS INJURED'].mean()
```

```python
bins = range(0, 50, 5)
pd.cut(data['NUMBER OF PERSONS INJURED'], bins = bins, include_lowest = True).value_counts(dropna = False).sort_index()
```

### Looking at how the data distributes temporally

```python
date_col = pd.to_datetime(data['DATE'], format = '%m/%d/%Y').rename('DATE2')
```

```python
data = pd.concat([date_col, data], axis = 1)
```

```python
data2 = data.groupby('YEAR')['NUMBER OF PERSONS INJURED'].agg('sum')
data2 = data2.reset_index()
sns.barplot(data2['YEAR'], data2['NUMBER OF PERSONS INJURED'], palette=sns.color_palette(n_colors=1))
```
